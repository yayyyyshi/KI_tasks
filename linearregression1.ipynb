```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
np.random.seed(0)
X = 10 * np.random.rand(100, 1)
y = 24 + 3 * X + np.random.randn(100, 1)
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
X_new = np.array([[0], [10]])  # 10 end-points for line
y_pred = model.predict(X_new)
```

```python
plt.figure(figsize=(10, 5))
plt.scatter(X, y, color="blue", label="Data points")
plt.plot(X_new, y_pred, color="red", linewidth=2, label="Regression line")
plt.title("Linear Regression Visualization")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
print(f"Intercept (θ₀): {model.intercept_[0]:.2f}")
print(f"Slope     (θ₁): {model.coef_[0][0]:.2f}")
```
